```python
# Import the necessary libraries to process data:
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib as mpl
# Import the required libraries to perform operations on time series:
from datetime import datetime, timedelta, date
```

```python
### Question 1A:
### A) Retrieve data for a stock (e.g., AAPL) for the past year using yfinance.
### B) Compute and plot the moving average

def fetch_data(ticker="AAPL") -> pd.DataFrame:
    # Define the time interval:
    endDate = date.today() # Ex: "2024-12-02"
    startDate = date.today().replace(month=1, day=1) # Ex: "2024-01-01"
        
    try: # Attempt to fetch the data from the yfinance library:
        df = yf.download(ticker, start=startDate, end=endDate)

    except Exception as e: # Case where an internal error => return an empty DataFrame:
        print(f"Error couldn't fetch data for {ticker}: {str(e)}")
        return pd.DataFrame()

    # Add the ticker symbol as a column:
    df["Ticker"] = ticker
    
    # Flatten the MultiIndex:
    if isinstance(df.columns, pd.MultiIndex):
        # Only use the first level of the MultiIndex
        df.columns = df.columns.get_level_values(0)

    # Reset the index to insure the Date is a column:
    df = df.reset_index()

    # Drop the volume column:
    df.drop(columns=["Volume"], inplace=True)
    
    return df

raw_data = fetch_data()
print(raw_data)
```

```python
### Question 1B:
### B) Compute and plot the 5 day moving average
def compute_moving_average(df: pd.DataFrame) -> pd.DataFrame:
    # Compute the moving average based on the nbr of days worth of data:
    df["5-days SMA"] = df["Close"].rolling(window=50).mean()
    # Remove nan values:
    df.dropna(axis=0, subset=["5-days SMA"], inplace=True)
    return df

sma = compute_moving_average(raw_data)
print(sma)


import matplotlib.pyplot as plt
def plot_moving_average(data, sma_column="5-days SMA", price_column="Close", date_column="Date"):
    # Define the size of the plot:
    plt.figure(figsize=(14,7))
    # Plot the stock price:
    plt.plot(data[date_column], data[price_column], label="Stock Price", color="blue", alpha=0.6)
    # Plot the moving average:
    plt.plot(data[date_column], data[sma_column], label=f"{sma_column}", color="orange", alpha=0.8)
    # Give a title:
    plt.title(f"Stock Price and {sma_column}", fontsize=16)
    plt.xlabel("Date", fontsize=12)
    plt.ylabel("Price", fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Rotate date labels for better visibility:
    plt.xticks(rotation=45)
    plt.show()

plot_moving_average(sma)
```
